# Handle missing values

```python
# Check if 'Cabin' column exists before trying to drop it
if 'Cabin' in train.columns:
    train = train.drop(columns=['Cabin'])

# Impute 'Age' with the median and assign back to the original DataFrame
train['Age'] = train['Age'].fillna(train['Age'].median())

# Impute missing values in 'Embarked' with the mode and assign back to the original DataFrame
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
```

# Print the Cleaned DataFrame

```python
# Handle outliers in 'Fare' using the IQR method
Q1 = train['Fare'].quantile(0.25)
Q3 = train['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers and filter them out
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Removing outliers from 'Fare'
train_cleaned = train[(train['Fare'] >= lower_bound) & (train['Fare'] <= upper_bound)]

# Display the cleaned DataFrame
print(train_cleaned.head())

```

```python
# Handle outliers in 'Fare' using the IQR method
Q1 = train['Fare'].quantile(0.25)
Q3 = train['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers and filter them out
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Removing outliers from 'Fare'
train_cleaned = train[(train['Fare'] >= lower_bound) & (train['Fare'] <= upper_bound)]

# Save the cleaned DataFrame to a CSV file
train_cleaned.to_csv('cleaned_train_dataset.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Count missing values per column
missing_counts = train.isnull().sum()

# Plot the missing values as a bar chart
plt.figure(figsize=(10, 6))
missing_counts.plot(kind='bar')
plt.ylabel('Number of Missing Values')
plt.show()
```

```python
# Count missing values per column in the train dataset
missing_counts = train.isnull().sum()

# Filter out columns with no missing values
missing_counts = missing_counts[missing_counts > 0]

# Check if there are any missing values before plotting
if not missing_counts.empty:
    # Plot the missing values as a bar chart
    plt.figure(figsize=(10, 6))
    missing_counts.plot(kind='bar', color='skyblue')
    plt.title('Missing Values per Column')
    plt.ylabel('Number of Missing Values')
    plt.xlabel('Columns')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No missing values in the dataset.")
```
