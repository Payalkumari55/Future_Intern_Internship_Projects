# Calculate Summary Statistics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Loading

```python
info = df.info()
info
```

```python
description = df.describe()
description
```

```python
# Calculate summary statistics (mean, median, mode, standard deviation)
summary_statistics = {
    "Mean": df.mean(numeric_only=True),
    "Median": df.median(numeric_only=True),
    "Mode": df.mode(numeric_only=True).iloc[0],  # Select the first mode
    "Standard Deviation": df.std(numeric_only=True)
}

summary_statistics
```

# Train

```python
train_df = pd.read_csv('F:/Future Intern/Task 2/titanic/train.csv')
```

```python
info = train_df.info()
info

# Describe the data
description = train_df.describe()
description

# Top 5
head = train_df.head()
head
```

```python
# Calculate summary statistics (mean, median, mode, standard deviation)
summary_statistics = {
    "Mean": train_df.mean(numeric_only=True),
    "Median": train_df.median(numeric_only=True),
    "Mode": train_df.mode(numeric_only=True).iloc[0],  # Select the first mode
    "Standard Deviation": train_df.std(numeric_only=True)
}

summary_statistics
```

# Visualization

```python
# Visualize the summary statistics using bar charts

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Mean
mean_values = train_df.mean(numeric_only=True)
axs[0, 0].bar(mean_values.index, mean_values.values, color='lightblue')
axs[0, 0].set_title('Mean of Numeric Columns')
axs[0, 0].set_xticklabels(mean_values.index, rotation=45)

# Median
median_values = train_df.median(numeric_only=True)
axs[0, 1].bar(median_values.index, median_values.values, color='lightgreen')
axs[0, 1].set_title('Median of Numeric Columns')
axs[0, 1].set_xticklabels(median_values.index, rotation=45)

# Mode
mode_values = train_df.mode(numeric_only=True).iloc[0]
axs[1, 0].bar(mode_values.index, mode_values.values, color='lightcoral')
axs[1, 0].set_title('Mode of Numeric Columns')
axs[1, 0].set_xticklabels(mode_values.index, rotation=45)

# Standard Deviation
std_values = train_df.std(numeric_only=True)
axs[1, 1].bar(std_values.index, std_values.values, color='lightgoldenrodyellow')
axs[1, 1].set_title('Standard Deviation of Numeric Columns')
axs[1, 1].set_xticklabels(std_values.index, rotation=45)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
